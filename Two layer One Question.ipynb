### Try to learn the answer to a question based on all other answers. Using a two layer NN

```python
import pandas as pd
from sklearn.model_selection import train_test_split
# one_hot_data = pd.read_pickle("./one_hot.bz2")

non_question_data_one_hot = pd.read_pickle("./non_question_data_one_hot.bz2") #input
question_data_one_hot = pd.read_pickle("./question_data_one_hot.bz2") #out labels

#Leave one question only and get rid of lines with Nan
#TODO get questions in non one hot, pick one column and convert to one hot here
non_question_data_one_hot = non_question_data_one_hot[question_data_one_hot.q9688_Nan == 0]
question_data_one_hot = question_data_one_hot[question_data_one_hot.q9688_Nan == 0]
question_data_one_hot = question_data_one_hot[['q9688_No','q9688_Yes','q9688_Yes, but only soft stuff like marijuana']]
```

```python
print question_data_one_hot.shape
print question_data_one_hot.shape
question_data_one_hot[:5]
```

```python
from keras.models import Model
from keras.layers import Input, Dense, Lambda
from keras import layers
from keras.initializers import Ones, Zeros
import numpy as np

# fix random seed for reproducibility
np.random.seed(7)

input_layer = Input(shape=(non_question_data_one_hot.shape[1],))
hidden_layer = Dense(non_question_data_one_hot.shape[1], activation='relu')(input_layer)
output_layer = Dense(question_data_one_hot.shape[1], activation='softmax')(hidden_layer)

model = Model(inputs=input_layer, outputs=output_layer)

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint
filepath="weights-two-layer-one-question.hdf5" #-{epoch:02d}-{val_acc:.2f}
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', save_best_only=True, mode='max')
callbacks_list = [checkpoint]

# print question_data_one_hot[[out_titles]]
model.fit(
          non_question_data_one_hot.values, 
          question_data_one_hot.values, 
          epochs=10000, batch_size=1024*64, validation_split=0.05,
          callbacks=callbacks_list)
```
